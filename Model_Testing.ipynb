The client will give a file with n number of input variables along with values we have to use the ML model we built and we have to predict the output for the entire data and we have convert it to a CSV file and download it!

Then we have to share that downloaded client output file with the client!


```python
import pickle
```

```python
#uploaading the downloaded ML model
with open('/content/housing_model_2','rb') as f:
  reg = pickle.load(f)
```

```python
#Testing whether our ML is working or not by giving own input
reg.predict([[1850,3,20]])
```

```python
#uploading the client input csv file
import pandas as pd
df = pd.read_csv("/content/client_file.csv")
df
```

```python
predicted_price = reg.predict(df)
predicted_price
```

```python
print(type(predicted_price))
```

```python
#crating a new column in the existing dataframe (df)
df["price"] = predicted_price
print(df)
```

```python
#converting the dataframe into csv file
df.to_csv("output.csv",index = False)
```
