

---



#Build **ML** Model

---



```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
#Train the model using fit function
reg.fit(X,y)
```

```python
reg.predict([[1850]])
```

```python
reg.score(X,y)
```

#**y = mX+b**

m --> coef of input variable X

X --> Input Variablr(Independent Variable)

y --> output Variable (Dependent Variablr)

```python
m = reg.coef_
m
```

```python
b = reg.intercept_
b
```

```python
input_X = 1850
output_price= m*input_X + b
output_price
```

```python
plt.scatter(X,y,color="green")
ytest = m*X + b
plt.plot(X,ytest,color='red')
```
